```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()

x = np.linspace(-2, 1, 100)
y = 10*x + 2
plt.plot(x, y, color='black', label='y=β0+β1x', linewidth=2, linestyle='-')
plt.plot((2,3),(4,2))

plt.axhline(y=0, color='black', linestyle='-', linewidth=1)
plt.axvline(x=0, color='black', linestyle='-', linewidth=1)
ax.hlines(y=2, xmin=-2, xmax=0, linewidth=1, color='black', linestyle='--')

plt.yticks([2], ['β0'], fontsize=14)
plt.xticks([0], ['0'], fontsize=14)
plt.title('Straight line equation', fontsize=16)
plt.xlim([-0.3, 0.45])
plt.ylim([-0.4, 6])

# plt.text(0.3, 1, 'y = β0 + β1x', fontsize = 14, 
#          bbox = dict(facecolor = 'grey', alpha = 0.5))
plt.legend(prop={'size': 12})
plt.tight_layout()
plt.savefig('../plots/Straight_line_equation_example.png')
plt.show()
```
