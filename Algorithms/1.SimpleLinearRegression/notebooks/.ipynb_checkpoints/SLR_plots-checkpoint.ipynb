# Plots for Simple Linear Regression Chapter

Import required libraries.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

Load dataset as a pandas frame.

```python
data_raw = pd.read_csv('../datasets/Greece - Agriculture and Rural Development/agriculture-and-rural-development_grc.csv')
```

Define and extract the variable names to use in analysis. Drop rows that contain NaN values and reset the index.

```python
variable_x = "Crop production index (2014-2016 = 100)"
variable_y = "Cereal production (metric tons)"
data = data_raw[[variable_x, variable_y]].dropna().reset_index(drop=True)
```

Transform variable to numpy arrays and set the type as 'np.float64'.

```python
X, y = np.array(data[variable_x], dtype=np.float64), np.array(data[variable_y], dtype=np.float64)
```

Create a Linear Regression model and train it using the fit() function. We use the reshape() function to respahe the training vector expects a 2D array whereas the target vector expects a 1D array. For more information see the [documentation](https://scikit-learn.org/stable/modules/generated/sklearn.naive_bayes.GaussianNB.html#sklearn.naive_bayes.GaussianNB.fit). 

```python
model = LinearRegression().fit(X.reshape(-1, 1), y)
```

Now, we use the predict() function to make a prediction for our training data. Note that using the same data in both training and testing phases is wrong, but in this case we just want some predictions for plotting.

```python
pred = model.predict(X.reshape(-1, 1))
```

We can use the preditions and the initial data to plot various plots for our project.

## Regression line & scatter plot

```python
# Plot the regression line

plt.figure() 
plt.scatter(X, y, color='red')
plt.plot(X, pred, color="black", linewidth=2)

plt.title('Residuals - Cereal production', fontsize=16)
plt.xlabel(variable_x, fontsize=14)
plt.ylabel(variable_y, fontsize=14)


plt.tight_layout()
plt.savefig('../plots/LSE_fit_Cereal_production.png')
plt.show()
```

## Regression line & scatter plot & residuals

```python
# PLot 

plt.figure() 
plt.scatter(X, y, color='red')
plt.plot(X, pred, color="black", linewidth=2)

plt.title('Residuals - Cereal production', fontsize=16)
plt.xlabel(variable_x, fontsize=14)
plt.ylabel(variable_y, fontsize=14)

plt.vlines(x=X, ymin=y, ymax = pred, 
           colors = 'black', 
           label = 'vline_multiple - partial height', linewidth=0.5)

plt.tight_layout()
plt.savefig('../plots/residuals_LSE_fit_Cereal_production.png')
plt.show()
```

## Cover 

```python
# Make scatter plot
f = plt.figure() 
plt.scatter(X, y, color='red', s=50)
plt.plot(X, pred, color="black", linewidth=3)

plt.xticks([])
plt.yticks([])
f.set_figwidth(6)
f.set_figheight(5)


plt.vlines(x=X, ymin=y, ymax = pred, 
           colors = 'black', 
           label = 'vline_multiple - partial height', linewidth=1)

plt.tight_layout()
plt.savefig('../plots/cover.png', dpi=200)
plt.show()
```

## Random lines & scatter plot

```python
###### Make scatter plot
plt.tight_layout()
plt.scatter(X, y, color='red')
plt.title('Straight lines - Cereal production', fontsize=16)
plt.xlabel(variable_x, fontsize=14)
plt.ylabel(variable_y, fontsize=14)
p1 = 17
p2 = 19
p3 = 32
p4 = 50
plt.axline(xy1=(X[p1], y[p1]), xy2=(X[p2], y[p2]), color='black', linewidth=2)
plt.axline(xy1=(X[p3], y[p3]), xy2=(X[p4], y[p4]), color='black', linewidth=2)

plt.tight_layout()
plt.savefig('../plots/random_lines_Cereal_production.png')
plt.show()
```
