```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()

x = np.linspace(0, 30, 30)
y = 1000*x
yy = 1000*x + 5000
plt.plot(x, y, color='black', label='y=1000x', linestyle='--')
plt.plot(x, yy, color='black', label='y=1000x+5000')

plt.title('Savings calculation', fontsize = 16)
plt.xlabel('Years', fontsize = 14)
plt.ylabel('Amount', fontsize = 14)

plt.xlim([0, 32])
plt.ylim([0, 40000])

plt.tight_layout()
plt.legend(prop={'size': 12})
plt.savefig('../plots/Savings_example_plot.png')
plt.show()
```
