```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data_raw = pd.read_csv('./agriculture-and-rural-development_grc.csv')
```

```python
data_raw.head()
```

```python
data = data_raw.pivot(index='Year', columns="Indicator Name", values="Value")
data
```

```python
data.columns
```

```python
variable_1 = "Crop production index (2014-2016 = 100)"
variable_2 = "Cereal production (metric tons)"
temp_data = data[[variable_1, variable_2]].dropna()
x, y = np.array(temp_data[variable_1]), np.array(temp_data[variable_2])
```

```python
y
```

# Straight line equation

```python
###### Make scatter plot
plt.tight_layout()
plt.scatter(x, y, color='blue')
plt.title('Straight lines - Cereal production', fontsize=16)
plt.xlabel(variable_1, fontsize=14)
plt.ylabel(variable_2, fontsize=14)
p1 = 17
p2 = 19
p3 = 32
p4 = 50
plt.axline(xy1=(x[p1], y[p1]), xy2=(x[p2], y[p2]), color='r', linewidth=2)
plt.axline(xy1=(x[p3], y[p3]), xy2=(x[p4], y[p4]), color='r', linewidth=2)

plt.tight_layout()
plt.savefig('plots/StraightLines_for_Regression_Cereal_production.png')
plt.show()
```

# Linear Regression

```python
x = x.reshape(-1, 1)
```

```python
model = LinearRegression().fit(x, y)
```

```python
pred = model.predict(x)
```

```python
plt.style.use('default')
```

```python
# Make scatter plot
#plt.figure() 
plt.scatter(x, y, color='red')
plt.plot(x, pred, color="black", linewidth=2)

plt.title('Linear regression - Cereal production', fontsize=16)
plt.xlabel(variable_1, fontsize=14)
plt.ylabel(variable_2, fontsize=14)

plt.tight_layout()
plt.savefig('plots/SimpleLinearRegression_LSE_fit_Cereal_production.png')
plt.show()
```

# Linear Regression - Residuals

```python
# Make scatter plot
plt.figure() 
plt.scatter(x, y, color='red')
plt.plot(x, pred, color="black", linewidth=2)

plt.title('Residuals - Cereal production', fontsize=16)
plt.xlabel(variable_1, fontsize=14)
plt.ylabel(variable_2, fontsize=14)

plt.vlines(x = x, ymin=y, ymax = pred, 
           colors = 'black', 
           label = 'vline_multiple - partial height', linewidth=0.5)

plt.tight_layout()
plt.savefig('plots/SimpleLinearRegression_residuals_LSE_fit_Cereal_production.png')
plt.show()
```
